### Setup modules

```python
import os
import sys
sys.path.append(os.path.join(os.path.dirname("__file__"), "../../Modules"))

from ImageUtilities import imageReadL, plotImageL, createImageL 
from PrintUtilities import printImageRangeL

# Iteration
from timeit import itertools
```

### Parameters

```python
# Increase brightness
brightDelta = 80
# Image range to print
printRange = [0,10]
```

### Read an image

```python
imageName = "Zebra.png"
pathToDir = os.path.join(os.path.dirname("__file__"), "../../Images/Chapter1/Input/")
inputImage, width, height  = imageReadL(pathToDir + imageName)
```

### Create output with chnaged intensty

```python
outputImage = createImageL(width, height)

# Set the pixels in the output image
for x,y in itertools.product(range(0, width), range(0, height)):  
    outValue = int(inputImage[y,x]) + brightDelta 
    if outValue < 255:
        outputImage[y,x] = outValue 
    else:
        outputImage[y,x] = 255
```

### Show images

```python
plotImageL(inputImage)
plotImageL(outputImage)
```

### Print an image range

```python
printImageRangeL(inputImage, printRange, printRange)
printImageRangeL(outputImage, printRange, printRange)
```
